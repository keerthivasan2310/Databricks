```python
from delta.tables import DeltaTable
def id_change(path,old_id,new_id):
  delt = DeltaTable.forPath(spark,path)
  delt.update(condition=f"id={old_id}",set={"id":f"{new_id}"})
  display(delt.toDF())
```

```python
id_change("/Volumes/workspace/default/vol_1/delta",3,4)
```
