# standard connector


```python

df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("/Volumes/workspace/connector/lake/sales.csv")

df.show()
```

```python
df.filter(df.price > 300).show()
```

```python
df.show()
```

```python
df.write.saveAsTable("sales_table")
#spark.sql("DROP TABLE sales_table")
```

```python
spark.sql("SELECT * FROM sales_table").show()
```

```python
df.write.format("delta").mode("overwrite").save("/Volumes/workspace/connector/lake/sales")
```

```python
delta_df = spark.read.format("delta").load("/Volumes/workspace/connector/lake/sales")

delta_df.show()
```

## migrate to delta

```python
data = [("New York", 100), ("Los Angeles", 200), ("Chicago", 150)]
columns = ["city", "price"]
df = spark.createDataFrame(data, columns)

df.write.mode("overwrite").parquet("/Volumes/workspace/connector/lake/sales_parquet")
```
